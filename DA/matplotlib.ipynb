**Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

    Functional

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Basic')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g')

plt.subplot(1,2,2)
plt.plot(y,x, 'r')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'g')

plt.subplot(2,1,2)
plt.plot(y,x, 'r')
```

    Object Oriented

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.45,0.2,0.4,0.4])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.45,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('(X,Y)')

axes2.plot(y,x)
axes2.set_title('(Y,X)')
```

```python
fig = plt.figure()

axes = fig.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout() #to takeout overlapping
```

    Figure Size and DPI

```python
fig = plt.figure(figsize=(5,2.5))#,dpi=75)

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(6,5))

axes = fig.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('X-Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Y-X')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

plt.tight_layout()
```

```python
fig.savefig('my_fig.png',dpi=200)
```

```python
fig1 = plt.figure()

ax = fig1.add_axes([0,0,1,1])

ax.plot(x, x * 2, label='X Doubled')
ax.plot(x, x ** 2, label='X Squared')


ax.legend()
```

    Plot Appearance

```python
fig = plt.figure(figsize=(5,4))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='#8002D4',linewidth=2,alpha=0.5,linestyle='-',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='green' )

plt.show()
```

```python
fig = plt.figure(figsize=(5,4))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='#8002D4',linewidth=2,alpha=0.5,linestyle='-')

axes.set_xlim([0,2])
axes.set_ylim([0,3])

plt.show()
```

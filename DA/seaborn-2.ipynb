    LM plots
An LM plot (also known as a linear model plot) in Python is a regression plot that visualizes the relationship between two variables along with a fitted regression line. It is part of the seaborn library and is a great tool for visualizing both the linear regression model and the data points that fit (or deviate from) the model.

```python
plt.figure()
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')

plt.title("LM Plot Total_bill Vs. Tip")
plt.show()
```

```python
plt.figure()
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['o','v'],
            scatter_kws={'s':75})

plt.show()
```

```python
plt.figure()
sns.lmplot(data=tips,x='total_bill',y='tip',col='sex',row='time')

plt.show()
```

```python
plt.figure()
sns.lmplot(data=tips,x='total_bill',y='tip',col='day',hue='sex',
           aspect= 0.6)

plt.show()
```

    ScatterPlot
Actually, there is a scatterplot function in seaborn. In recent versions of seaborn, scatterplot is the preferred way to create scatter plots because it offers more flexibility and customization compared to lmplot or older methods like regplot.

Key Differences Between lmplot and scatterplot:

**scatterplot():**
It's purely for scatter plots. It does not fit any regression line to the data.
Provides flexibility for color, size, style mappings, and customization.
Newer and more feature-rich compared to older functions like lmplot.

**LMplot():**
Automatically fits and shows a linear regression line along with the scatter points.
Useful if you want both scatter points and a regression line in one plot.

```python

plt.figure(figsize=(8,6))
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',size='size',style='smoker')


plt.title("ScatterPlot Total_bill VS Tip")
plt.show()
```

```python
sns.set_style('ticks')
sns.countplot(data=tips,x='sex')
sns.despine()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=tips,x='smoker')
sns.despine(left=True,bottom=True)
plt.show()
```
